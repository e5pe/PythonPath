## Histogramas

Un histograma es una representación de una variable o de unos datos en forma de barras, donde la superficie de cada barra es proporcional a la frecuencia de los valores representados.

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import seaborn as sns
```

```python
%matplotlib inline
```

La línea anterior es para que aparezcan los gráficos en el notebook de *jupyter*.

```python
datos = np.random.randn(100)
datos
```

```python
mpl.pyplot.hist(datos)
```

```python
sns.distplot(datos)
```

```python
sns.distplot(datos, color="purple")
```

```python
mpl.pyplot.hist(datos, color="purple",alpha=0.3)
```

```python
mpl.pyplot.hist(datos, color="purple",alpha=0.9)
```

```python
datos2 = np.random.randn(80)
mpl.pyplot.hist(datos2,color="green",alpha=0.4)
```

```python
mpl.pyplot.hist(datos,color="green",alpha=0.4, bins=20,density=True)
mpl.pyplot.hist(datos2,color="purple",alpha=0.4, bins=20,density=True)
```

**bins** sirve para indicar el número de columnas

## Combinando estilos                                            

```python
datos = np.random.randn(100)
datos
```

```python
sns.distplot(datos)
```

Vamos a quitar el gráfico de barras y dejar sólo la curva.

```python
sns.distplot(datos,color="green",rug=False, hist=False)
```

```python
argumentos_curva = {'color':'green','label':'Curva'}
argumentos_histo = {'color':'grey','label':'Histograma'}
sns.distplot(datos, bins=25, kde_kws=argumentos_curva,hist_kws=argumentos_histo)
```

Así se combinan ambos estilos, de la curva y del histograma, con diccionarios.

Vamos a ver que el *seaborn* se maneja muy bien con el *pandas*.

```python
serie = pd.Series(datos)
```

```python
sns.distplot(serie, bins=25, color="green")
```

## Regresiones lineales

En estadística es un proceso estadístico para estimar las relaciones entre variables.

**Regresión lineal** para ver la relación de dependencia entre variables.

```python
propinas = sns.load_dataset('tips')
```

```python
propinas.head(10)
```

```python
sns.lmplot('total_bill','tip',propinas)
```

Compara la factura total con la propina.

`scatter_kws` va a ser la y line_kws la x

```python
sns.lmplot('total_bill','tip',propinas,scatter_kws={'color':'green'}, line_kws={'color':'blue'})
```

Sólo los puntos sin la línea:

```python
sns.lmplot('total_bill','tip',propinas,fit_reg=False)
```

`fit_reg=False` para que no salga la línea en una regresión lineal 

```python
propinas.head()
```

```python
propinas['porcentaje'] = 100 * propinas['tip'] / propinas['total_bill']
```

```python
propinas
```

```python
sns.lmplot('size','porcentaje',propinas)
```

```python
sns.lmplot('total_bill','porcentaje',propinas,hue='sex',markers=['x','o'])
```

```python
sns.lmplot('total_bill','porcentaje',propinas,hue='day')
```

## Mapas de calor

*heatmap* o **mapa de calor**: Es una representación gráfica de datos donde los valores individuales contenidos en una matriz se representan como colores.

```python
vuelos = sns.load_dataset('flights')
```

```python
vuelos.head(10)
```

```python
vuelos = vuelos.pivot('month','year','passengers')
vuelos
```

Para hacer un mapa de calor necesitamos una matriz con filas y columnas.

Los parámetros de *pivot* son param1 -> filas, param2 -> columnas, param3 -> datos.

```python
sns.heatmap(vuelos)
```

```python
sns.heatmap(vuelos,annot=True,fmt='d')
```

```python
valor_central = vuelos.loc['May'][1956]
valor_central
```

```python
sns.heatmap(vuelos,center=valor_central, annot=True, fmt='d')
```

```python
sns.heatmap(vuelos,center=valor_central, annot=True, fmt='d', cbar_kws={'orientation':'horizontal'})
```
